```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
haberman=pd.read_csv(r"C:\Users\shyam\OneDrive\MACHINE LEARNING PROGRAMS\haberman.csv")
```

```python
haberman.columns=['Patient Age','Year of Operation','Number of positive axillary nodes','Survival status(1 or 2)']
```

haberman

```python
haberman.head(10)
```

```python
s = haberman.values[:,0:3]
```

```python
s
```

```python
scr = StandardScaler()
```

```python
x = scr.fit_transform(s)
```

```python
x
```

```python
y = haberman.values[:,3]
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,random_state=1,test_size=0.2)
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

```python
log_reg = LogisticRegression(random_state=0)
```

```python
log_reg.fit(xtrain,ytrain)
ypre = log_reg.predict(xtest)
```

```python
print(confusion_matrix(ytest,ypre))
```

```python
print(classification_report(ytest,ypre))
```

```python
print(accuracy_score(ytest,ypre))
score = accuracy_score(ytest,ypre)*100
score
```

```python
print('THE ACCURACY SCORE FOR HABERMAN DATA BY LOGISTIC REGRESSION ALGORITHM IS:',score)
```
